```python
import os

os.makedirs("dataset/images/train", exist_ok=True)
os.makedirs("dataset/labels/train", exist_ok=True)

print("Folders created successfully!")
```

```python
import shutil
import glob

images = glob.glob("../dataset/PennFudanPed/PNGImages/*.png")

for img in images:
    shutil.copy(img, "dataset/images/train")

print("Images copied successfully!")
```

```python
import os
import numpy as np
from PIL import Image

mask_dir = "../dataset/PennFudanPed/PedMasks"
label_dir = "dataset/labels/train"

masks = os.listdir(mask_dir)

for mask_name in masks:

    mask_path = os.path.join(mask_dir, mask_name)

    mask = np.array(Image.open(mask_path))

    h, w = mask.shape

    ids = np.unique(mask)

    ids = ids[ids != 0]

    label_file = open(
        os.path.join(
            label_dir,
            mask_name.replace("_mask.png", ".txt")
        ),
        "w"
    )

    for obj_id in ids:

        pos = np.where(mask == obj_id)

        xmin = np.min(pos[1])
        xmax = np.max(pos[1])

        ymin = np.min(pos[0])
        ymax = np.max(pos[0])

        x_center = ((xmin + xmax) / 2) / w
        y_center = ((ymin + ymax) / 2) / h

        width = (xmax - xmin) / w
        height = (ymax - ymin) / h

        label_file.write(
            f"0 {x_center} {y_center} {width} {height}\n"
        )

    label_file.close()

print("YOLO labels created successfully!")
```
